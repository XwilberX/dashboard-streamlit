```
import pandas as pd
from scipy import  stats
import numpy as np
```

```
data = pd.read_csv('Datos_2016_2020_no_outlier.csv')
data = data.drop('Unnamed: 0', 1)
data.head()
```

```
# Normalizando fechas
data['date'] = pd.to_datetime(data['date'], unit='ns')
data.head()
```

```
# Funcion para media truncada al 10% de una columna x
def trim_mean_10(x):
    return stats.trim_mean(x, 0.1)
```

```
# Funcion para calcular el percentil n de una columna x
def percentile(n):
    def percentile_(x):
        return x.quantile(n)
    return percentile_
```

```
# esto para todos? Rh, Tmp, Wdr, Wsp
# Calculando los estimados de locacion y variabilidad
data.groupby('id_station').PM2_5.agg(
    media='mean',
    mediana='median',
    media_truncada=trim_mean_10,
    desv_estandar='std',
    minimo='min',
    percentile_25=percentile(0.25),
    percentile_50=percentile(0.5),
    percentile_75=percentile(0.75),
    maximo='max',
    rango=np.ptp, # max - min
    IQR=stats.iqr
)
```
