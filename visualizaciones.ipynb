```python
# Se cargan las librerias necesarias
import plotly.express as px
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# cargamos nuestro dataset
data = pd.read_csv('Datos_2016_2020_no_outlier.csv')
data = data.drop('Unnamed: 0', 1)
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data
```

```python
# df_original.groupby(pd.Grouper(key='Date',freq='M')).sum()

data['month'] = data['date'].dt.month
data['year'] = data['date'].dt.year

# data = data.groupby(pd.Grouper(key='date',freq='M')).mean()

data
```

```python
data.dtypes
```

```python
fig = px.treemap(data, path=['year', 'month', 'id_station'], values='PM2_5', color='PM2_5')
fig.show()
```
